```python
url = "https://en.wikipedia.org/wiki/Lists_of_Telugu-language_films";

headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
# Send an HTTP GET request to the URL
response = requests.get(url,headers=headers)
print(response);
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup

     soup = BeautifulSoup(response.content, 'html.parser')

     book_titles = soup.find_all('div', class_='mw-heading mw-heading2')

     # Movie Years
    # Extract and print the text of each title
     liststrings =[];
     for title_tag in book_titles:
        #litags = title_tag.find_all_next("li")
        #print(type(title_tag.h2.text[:-1]))
        #if title_tag.h2.text[:-1].isdigit():
         #num = int(title_tag.h2.text[:-1])
        #print(num)
        #if type(title_tag.h2.text[:-1]) is int:
           # print(title_tag.h2.text[:-1])
        all_ultags = title_tag.find_next('ul')
         #firstelement = all_ultags.find_all('li')[0]
         #all_ultags.find_all('li')

        for li_tag in all_ultags.find_all('li'):
             pattern = r"List of Telugu films of (\d{4})"
             #if(li_tag.a.text.find("List of Telugu films of") != -1):
             match = re.search(pattern, li_tag.a.text)
             if match:
               liststrings.append({"year": match.group() , "title":li_tag.a.text, "url":li_tag.a['href']})
     for year in liststrings[3:]:
         yearwiseMovies(year)
       #print(year)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")



```

```python
url = "https://en.wikipedia.org/wiki/List_of_Telugu_films_of_the_1930s";

headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
# Send an HTTP GET request to the URL
response = requests.get(url,headers=headers)
print(response);
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup

     soup = BeautifulSoup(response.content, 'html.parser')

     table = soup.find('table', class_='wikitable')

     headers = [th.text.strip() for th in table.find_all('th')]

      # extract table body
     rows = []

     for row in table.find_all('tr')[1:]:
       if len(row.find_all('td')) == 1:
        year = row.find_all('td')[0].text.strip()
       #print (year)
       if len(row.find_all('td')) > 1 :
        cells = [td.text.strip() for td in row.find_all('td')]
        cells.append(year)
        rows.append(cells)

# log data
print("Headers:", headers)
print("table_body:")
rows.insert(0,['Title', 'Director', 'Production', 'Cast', 'Release','Year'])
df = pd.DataFrame(rows)

df.to_csv('1930_to_1940_movies.csv', index=True,header=False)
print("DataFrame saved to 'output_file.csv' in the Colab session.")
#else:
#print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
